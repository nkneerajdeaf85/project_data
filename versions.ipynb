```python
df_ver_dump = pd.read_sql_query(
    """
    WITH new_dump AS (
        SELECT 
            server_id, 
            server_name, 
            CONVERT(dump USING utf8) AS dump,
            timestamp
        FROM versions
    )
    SELECT 
        server_id, 
        server_name,
        dump, 
        MAX(CASE 
            WHEN dump LIKE '%MySQL version:%' THEN 
                SUBSTRING_INDEX(SUBSTRING_INDEX(dump, 'MySQL version:', -1), '\n', 1)
            END) AS MySQL_Version,
        MAX(CASE 
            WHEN dump LIKE '%PHP version:%' THEN 
                SUBSTRING_INDEX(SUBSTRING_INDEX(dump, 'PHP version:', -1), '\n', 1)
            END) AS PHP_Version,
        MAX(CASE 
            WHEN dump LIKE '%Node.js version:%' THEN 
                SUBSTRING_INDEX(SUBSTRING_INDEX(dump, 'Node.js version:', -1), '\n', 1)
            END) AS Nodejs_Version,
        MAX(CASE 
            WHEN dump LIKE '%MariaDB version:%' THEN 
                SUBSTRING_INDEX(SUBSTRING_INDEX(dump, 'MariaDB version:', -1), '\n', 1)
            END) AS MariaDB_Version,
        MAX(CASE 
            WHEN dump LIKE '%Redis version:%' THEN 
                SUBSTRING_INDEX(SUBSTRING_INDEX(dump, 'Redis version:', -1), '\n', 1)
            END) AS Redis_Version,
        MAX(timestamp) AS latest_timestamp
    FROM new_dump
    GROUP BY server_id, server_name, dump;
    """,
    conn
)

df_ver_dump
```
